````python
#The number of jobs that are created per month
pd.read_sql(
    """
    select strftime('%m', created_at) as Month,count(role_id) as number_of_jobs
    from jobs
    group by Month
    """, db_conn
)
````

````python
#number of jobs but specified per role per month
pd.read_sql(
    """
    select strftime('%Y', created_at) as Year, strftime('%m', created_at) as Month,r.title,count(j.role_id)
    from jobs j 
    join roles r on r.id=j.role_id
    group by Year,Month,r.title
    """, db_conn
)
````

````python
#% of bookings over posted jobs  (over time) ------
pd.read_sql(
    """
    select strftime('%Y %m', created_at) as YearMonth, r.title,count(j.role_id) as number_of_jobs_posted,count(j.booked_worker) as number_of_bookings, CAST(ROUND(count(j.booked_worker) * 100.0/count(j.role_id),2) as decimal(5,2)) as Percentage_Of_Bookings
    from jobs j 
    join roles r on r.id=j.role_id
    group by YearMonth,r.title
    """, db_conn
)
````

````python
# average payrate based on roles
pd.read_sql(
    """
      select r.title,avg(j.pay_rate),count(j.role_id)
      from jobs j
      join roles r on r.id=j.role_id
      group by r.title
    """, db_conn
)
````

````python
#Number of employers with certain average rating 
pd.read_sql(
    """
      with CTE_AvgRatingToEmployer(EmployerBrand,AvgRating) as
      (
      select e.brand,round(avg(j.rating_to_employer),0)
      from jobs j
      join employers e on e.id=j.employer_id
   
      group by 1
      order by 2 desc
      )
      select AvgRating, count(EmployerBrand)
      from CTE_AvgRatingToEmployer
      group by 1

    """, db_conn
)
````

````python
#8 average rating to employers 
pd.read_sql(
    """
      select e.brand,avg(j.rating_to_employer)
      from jobs j
      join employers e on e.id=j.employer_id
      group by e.brand
    """, db_conn
)
````

````python
#11 average rating to each workers
pd.read_sql(
    """
      select w.first_name,avg(j.rating_to_worker)
      from jobs j
      join workers w on w.id=j.booked_worker
      group by w.first_name
    """, db_conn
)
````

````python
# 12 Average rating to worker for each role
pd.read_sql(
    """
      select w.first_name || w.last_name as WorkerName, r.title, avg(rating_to_worker)
      from jobs j
      join workers w on w.id=j.booked_worker
      join roles r on r.id=j.role_id
      group by WorkerName,r.title
    """, db_conn
)
````

````python
# Average rating to worker for each role per month
pd.read_sql(
    """
      select strftime('%Y %m', j.created_at) as YearMonth,r.title as Role_Title,avg(j.rating_to_worker) as Avg_Rating_to_Worker
      from jobs j
      join roles r on r.id=j.role_id
      group by YearMonth,r.title
    """, db_conn
)
````

````python
# Using CTE, avg rating to workers per role per month
pd.read_sql(
    """
      with cte_workers_roles_ratings(WorkerName,Role,AvgRating,YearMonth) as 
      (
      select w.first_name || w.last_name,r.title,
      avg(j.rating_to_worker),strftime('%Y %m', j.created_at)
      from 
      jobs j
      join workers w on j.booked_worker=w.id
      join roles r on j.role_id=r.id
      group by 1,2
      )

      select 
      WorkerName,
      Role,
      AvgRating,
      YearMonth
      From 
      cte_workers_roles_ratings
    """, db_conn
)

````

````python
#Total no of workers with certain rating per month
#avg no of workers with certain rating per month
pd.read_sql(
    """
      with cte_NoofWorkersRatingMonth(YearMonth,Rating,No_Of_Workers) as
      (
        select strftime('%Y %m',created_at),rating_to_worker,count(booked_worker)
        from jobs 
        group by 1,2
      )
      select YearMonth,Rating,No_Of_Workers
      from
      cte_NoofWorkersRatingMonth
    """, db_conn
)
````

````python
#avg no of workers with certain rating per month
pd.read_sql(
    """
      with cte_NoofWorkersRatingMonth(YearMonth,booked_worker,rating_to_worker) as
      (
        select strftime('%Y %m',created_at),booked_worker,rating_to_worker
        from jobs 
        group by 1,2
      )
      select YearMonth,booked_worker,rating_to_worker
      from
      cte_NoofWorkersRatingMonth
    """, db_conn
)
````

````python
#Top 10 employers based on number of booked workers

pd.read_sql(
    """
      with CTE_AllEmployersByBookedWorkers(EmployerBrand,NumberOfBookedWorkers) as
      (
      select e.brand ,count(j.booked_worker)
      from jobs j
      join employers e on j.employer_id=e.id
      group by 1
      
      )
      select * from CTE_AllEmployersByBookedWorkers
      order by NumberOfBookedWorkers desc
      limit 10
      
    """, db_conn
)
````

````python
#Average  number of booked workers per employer

df = pd.read_sql(
    """
      with CTE_AllEmployersByBookedWorkers(EmployerBrand,NumberOfBookedWorkers) as
      (
      select e.brand,count(j.booked_worker)
      from jobs j
      join employers e on j.employer_id=e.id
      group by 1
      )
      select EmployerBrand,avg(NumberOfBookedWorkers) from CTE_AllEmployersByBookedWorkers
      group by 1
      order by 2 desc
      limit 10    
    """, db_conn
)
````

````python
#Employers growth over time 
pd.read_sql(
    """
      select strftime('%Y %m',e.created_at) as YearMonth,i.name,count(e.id) as CountOfEMployers
      from employers e
      join industries i on i.id=e.industry_id
      group by 1,2
    """, db_conn
)
````

````python
#Percentage of Employers growth over time per Industry
pd.read_sql(
    """
      with CTE_GrowthofEmployersOverTime(YearMonth,Industry,CountOfEMployers) as
      (
      select strftime('%Y %m',e.created_at) ,i.name,count(e.id)
      from employers e
      join industries i on i.id=e.industry_id
      group by 1,2
      ),
      total_employers(totalemployers) as
      (
        select count(id) from employers
      )
      
      select YearMonth,Industry,CAST(ROUND(CountOfEMployers* 100.0/totalemployers,2) as decimal(5,2)) as PercentageOfEmployers
      from CTE_GrowthofEmployersOverTime, total_employers
      group by 1,2

    """, db_conn
)
````

````python
#Percentage of Employers growth over time per Industry
pd.read_sql(
    """
      with CTE_GrowthofEmployersOverTime(YearMonth,CountOfEMployers) as
      (
      select strftime('%Y %m',e.created_at),count(e.id)
      from employers e
      join industries i on i.id=e.industry_id
      group by 1
      ),
      total_employers(totalemployers) as
      (
        select count(id) from employers
      )
      
      select YearMonth,Industry,CAST(ROUND(CountOfEMployers* 100.0/totalemployers,2) as decimal(5,2)) as PercentageOfEmployers
      from CTE_GrowthofEmployersOverTime, total_employers
      group by 1,2

    """, db_conn
)
````
